### KNN 알고리즘, Kmeans 
- KNN 지도학습 (회귀, 분류), Kmeans 비지도학습
### Kmeans 
- k개를 지정하여 평균을 통해 군집화한다는 뜻
- 비지도학습이라는게 정답을 준 것이 아니라 정답이 없는 데이터를 머신에게  패턴을 찾아달라고 하는 것 
- y 라벨이 없는 상태에서 데이터를 집어넣고 특성 군집화
- 군집화는 같은 패턴끼리 묶는다
- 작동원리가 어떤 식으로 진행하는지 : 평균
- 거리의 기반으로 작동한다 (유클리디안 거리)
- k-means의 k는 몇개의 클러스터를 할 것인가?
- k는 분석가가 초기값을 입력해야한다. 
- 초기값을 지정해 줘야함
- 단점 : 거리기반이기 때문에 비슷한 거리여도 클러스터링 함
- 거리 기반이라 이상치에 민감해짐
- 거리만이라 분산 구조는 반영 못함
- 데이터 양이 많아 반복 횟수가 많아지면 속도 느린 부분
- 몇개 군집으로 해야하는지 가이드 어렵
- 이해도가 쉽다. 알고리즘 간결. 거리 기반으로 평균으로 이해하면 되니까
